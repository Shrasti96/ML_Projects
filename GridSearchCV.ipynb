```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = sns.load_dataset('iris')
```

```python
df.head()
```

```python
df['species'].unique()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = df.drop('species', axis=1)
y = df['species']
```

```python
# train_test_split()      isme se niche vala code copy krna h
```

```python
X__train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
# model_knn = KNeighborsClassifier(n_neighbors=5)
model_knn = KNeighborsClassifier(n_neighbors=13)  #  if I am getting accuracy score 98% 100% then model is overfit
```

```python
model_knn.fit(X__train, y_train)
```

```python
model_knn.score(X_test, y_test)
```

```python
# JUst for check using SVM
```

```python
from sklearn.svm import SVC
```

```python
model_svm = SVC(gamma= 'auto')
```

```python
# model_svm = SVC(C = 30, kernel='rbf', gamma = 'auto')              # C is like panalty
# model_svm = SVC(C = 30, kernel='linear', gamma = 'auto')
# model_svm = SVC(C = 10, kernel='rbf', gamma = 'auto')
```

```python
model_svm.fit(X__train, y_train)
```

```python
model_svm.score(X_test, y_test)
```

```python
# Now lets use the grid search CV
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
classifier = GridSearchCV(model_svm, {
    'C':[1,10,20],
    'kernel':['rbf', 'linear']
}, cv=5, return_train_score=False)
```

```python
classifier.fit(X, y)
```

```python
classifier.cv_results_
```

```python
result = pd.DataFrame(classifier.cv_results_)
result
```

```python
result[['param_C', 'param_kernel', 'mean_test_score']]
```

```python
# Random search CV
```

```python
from sklearn.model_selection import RandomizedSearchCV
```

```python
classifier = RandomizedSearchCV(model_svm, {
    'C':[1,10,20, 30],
    'kernel':['rbf', 'linear']
}, cv=5, return_train_score=False, n_iter=4)
```

```python
classifier.fit(X, y)
```

```python
results = pd.DataFrame(classifier.cv_results_)
results
```

```python
results[['param_C', 'param_kernel','mean_test_score']]
```
